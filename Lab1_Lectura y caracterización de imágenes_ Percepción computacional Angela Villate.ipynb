### imagen 1: la bicicleta

```python
#se importan las librerías
import cv2
from matplotlib import pyplot as plt
from pylab import *
```

```python
# Cargar la imagen
imagen = cv2.imread('image1.png')

# Verificar si la carga de la imagen fue exitosa
if imagen is None:
    print("Error al cargar la imagen.")
else:
    print("Imagen cargada correctamente.")

# Mostrar la imagen
cv2.imshow('Imagen', imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convertir la imagen a formato RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar la imagen RGB
cv2.imshow('Imagen RGB', imagen_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Calcular el histograma
hist1 = cv2.calcHist([imagen_rgb], [0], None, [256], [0, 256])

# Mostrar el histograma
plt.plot(hist1, color='black')  # Corregir aquí, cambia 'hist' a 'hist4'
plt.title('Histograma de la Imagen')
plt.xlabel('Niveles de Intensidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Dividir la imagen en sus componentes de color (BGR)
blue_channel, green_channel, red_channel = cv2.split(imagen)

# Mostrar cada componente por separado
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Canal Azul')

plt.subplot(132)
plt.imshow(green_channel, cmap='Greens')
plt.title('Canal Verde')

plt.subplot(133)
plt.imshow(red_channel, cmap='Reds')
plt.title('Canal Rojo')

plt.show()
```

```python
# Cargar la imagen en escala de grises
imagen_gris = cv2.imread('image1.png', cv2.IMREAD_GRAYSCALE)

# Verificar si la carga de la imagen fue exitosa
if imagen_gris is None:
    print("Error al cargar la imagen en escala de grises.")
else:
    print("Imagen en escala de grises cargada correctamente.")

# Mostrar la imagen en escala de grises
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.show()
```

```python
 #Cargar la imagen en color
imagen_color = cv2.imread('image1.png')

# Verificar si la carga de la imagen fue exitosa
if imagen_color is None:
    print("Error al cargar la imagen.")
else:
    print("Imagen cargada correctamente.")

# Verificar si la imagen es en escala de grises
if len(imagen_color.shape) == 2:
    print("La imagen ya está en escala de grises.")
else:
    # Dividir la imagen en sus componentes de color (BGR)
    blue_channel, green_channel, red_channel = cv2.split(imagen_color)

    # Mostrar cada componente en escala de grises
    plt.figure(figsize=(12, 4))

    plt.subplot(131)
    plt.imshow(blue_channel, cmap='gray')
    plt.title('Canal Azul en Escala de Grises')

    plt.subplot(132)
    plt.imshow(green_channel, cmap='gray')
    plt.title('Canal Verde en Escala de Grises')

    plt.subplot(133)
    plt.imshow(red_channel, cmap='gray')
    plt.title('Canal Rojo en Escala de Grises')

    plt.show()
```
